```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Data provided by user
data = {
    'hour_of_day': ['12:00', '13:00', '14:00', '18:00', '17:00', '19:00', '16:00', '11:00', '15:00', '20:00', '21:00', '22:00', '23:00', '10:00', '09:00'],
    'total_quantity': [36083, 34438, 16228, 14731, 14026, 11730, 11387, 10531, 9755, 9291, 6800, 3613, 190, 43, 16],
    'avg_quantity': [11924] * 15,
    'total_revenue': [596045, 569168, 266062, 243094, 231967, 193382, 188810, 173168, 160831, 152649, 112461, 59375, 3121, 730, 332],
    'avg_revenue': [196746] * 15
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plotting Total Quantity by Hour of the Day with rotated x-axis labels and gridlines
plt.figure(figsize=(10, 6))
plt.bar(df['hour_of_day'], df['total_quantity'], color='steelblue')
plt.axhline(y=df['avg_quantity'][0], color='red', linestyle='--', label=f'Average Quantity: {df["avg_quantity"][0]}')
plt.xlabel('Hour of Day')
plt.ylabel('Total Quantity')
plt.title('Total Quantity by Hour of the Day')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, val in enumerate(df['total_quantity']):
    plt.text(i, val, str(val), ha='center', va='bottom', fontsize=9)
plt.show()

# Plotting Total Revenue by Hour of the Day with rotated x-axis labels, $ symbols for y-axis, and gridlines
plt.figure(figsize=(10, 6))
plt.bar(df['hour_of_day'], df['total_revenue'], color='steelblue')
plt.axhline(y=df['avg_revenue'][0], color='red', linestyle='--', label=f'Average Revenue: ${df["avg_revenue"][0]}')
plt.xlabel('Hour of Day')
plt.ylabel('Total Revenue ($)')
plt.title('Total Revenue by Hour of the Day')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Format y-axis with $ symbol
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))

# Adding values on the bars without $ symbol
for i, val in enumerate(df['total_revenue']):
    plt.text(i, val, f'{val}', ha='center', va='bottom', fontsize=9)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Data provided by user for day_of_week
data_week = {
    'day_of_week': ['Friday', 'Thursday', 'Saturday', 'Tuesday', 'Wednesday', 'Monday', 'Sunday'],
    'total_quantity': [27632, 27016, 26893, 26046, 25595, 23900, 21780],
    'total_revenue': [455140.90, 445174.40, 441206.35, 430556.60, 422855.50, 394880.40, 361382.80]
}

# Create a DataFrame for weekly data
df_week = pd.DataFrame(data_week)

# Plotting Total Quantity by Day of the Week with rounded average and gridlines
plt.figure(figsize=(10, 6))
plt.bar(df_week['day_of_week'], df_week['total_quantity'], color='steelblue')
avg_quantity_rounded = round(df_week['total_quantity'].mean())
plt.axhline(y=avg_quantity_rounded, color='red', linestyle='--', label=f'Average Quantity: {avg_quantity_rounded}')
plt.xlabel('Day of the Week')
plt.ylabel('Total Quantity')
plt.title('Total Quantity by Day of the Week')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, val in enumerate(df_week['total_quantity']):
    plt.text(i, val, str(val), ha='center', va='bottom', fontsize=9)
plt.show()

# Plotting Total Revenue by Day of the Week with rounded average, rounded values, $ symbols for y-axis, and gridlines
plt.figure(figsize=(10, 6))
plt.bar(df_week['day_of_week'], df_week['total_revenue'], color='steelblue')
avg_revenue_rounded = round(df_week['total_revenue'].mean())
plt.axhline(y=avg_revenue_rounded, color='red', linestyle='--', label=f'Average Revenue: ${avg_revenue_rounded:,}')
plt.xlabel('Day of the Week')
plt.ylabel('Total Revenue ($)')
plt.title('Total Revenue by Day of the Week')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Format y-axis with $ symbol
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))

# Adding rounded values on the bars without $ symbol
for i, val in enumerate(df_week['total_revenue']):
    plt.text(i, round(val), f'{round(val)}', ha='center', va='bottom', fontsize=9)

plt.show()
```
